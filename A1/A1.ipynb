Prove the following properties of Bernoulli Distribution.

<!-- BEGIN QUESTION -->

### Q4.1

$$\sum_{x=0}^1 p(x|\mu) = 1$$

_Points:_ 0.25

$$\sum_{x=0}^1 p(x|\mu) = 1$$
$$
\begin{align}
\sum_{x=0}^1 p(x|\mu) &= p(x=0|\mu) + p(x=1|\mu) = 1 \\
&= \mu^{x=0}\cdot(1-\mu)^{1-x=0} + \mu^{x=1}\cdot(1-\mu)^{1-x=1} = 1 \\
&= 1\cdot(1-\mu)^1 + \mu\cdot(1-\mu)^0 = 1 \\
&= 1\cdot(1-\mu) + \mu\cdot1 = 1\\
&= (1-\mu) + \mu = 1\\
&= 1 - \mu + \mu = 1\\
&= 1 = 1 \\
\end{align}
$$

<!-- END QUESTION -->

$$\mathbb{E}[x] = \mu$$

$$Var[x] = \mu(1-\mu)$$
